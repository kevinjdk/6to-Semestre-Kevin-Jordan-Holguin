### 1.- Crea una función en Python que calcule el mínimo común múltiplo de dos números.

```python
#MCM: ES EL NUMERO POSITIVI MAS PEQUEÑO QUE ES EL MULTIPO DE DOS NUMEROS

def mcm(x, y): #esta funcion tiene dos parametros en este caso x, y
    z = max(x, y) #iniciamos calculando el mayor de estos numeros 
    
    while True:
        if (z % x ==0) and (z % y ==0): # z debe ser divisible para x como para y
            return z
        z += 1 #si no se cumple esa condicion vamos a aunmentar a z una unidad para que se evalue el siguiente numero entero
        
print(mcm(3, 6))
print(mcm(5, 2))
print(mcm(17, 15))
```

### 2.- Crea una función en Python que calcule el máximo común divisor de dos números

```python
#MCD El numero mas grande que divide 2 numeros

def mcd(x, y): #recibe 2 parametros 
    mcd = 1 #creamos una variable asignandole 1, ya que 1 divide a cualquier numero
    if x % y == 0:
        return y #al hacer esto y seria el maximo comun divisor
    for k in range(int(y/2), 0, -1): #vamos a ir reduciendo y hasta que alcance 1
        if x % k == 0 and y % k == 0: #ahora comprobamos si k al calcular x e y es igual a 0
            mcd = k  #esto significa que k es el maximo comun divisor para x e y
            break #terminamos el ciclo
    return mcd


print(mcd(4, 8))
print(mcd(10, 20))
print(mcd(19, 29))
```

### 3.- Basándote en datos aleatorios  crea un gráfico de tarta (pie chart) con Mapplotlib.

```python
#Pie chart se grafica de forma antireloj

import matplotlib.pyplot as plt

labels = 'zapatos', 'camisa', 'gorra', 'jeans' #etiquetas
sizes = [30, 24, 26, 20] #los valores para el pie chart

print("Pie chart con numeros aleatorios")

figura_pieChart, ax1 = plt.subplots() #crea una figura y una cuadrícula de subtramas con una sola llamada.
ax1.pie(sizes, labels=labels, autopct='%1.1f%%') #llamamos los datos para que se grafiquen en el grafico.
ax1.axis('equal') #garantiza que el pie chart se dibuje como un circulo
plt.show()
```

Scatter Plot es usado en el estudio de la asociacion de dos variables de intervalo, donde ek grafico se muestra los valores de una variable sobre otra, para comprobar si es lineal cuadratica o ayudarnos para identificar patrones o clusters en los datos.
